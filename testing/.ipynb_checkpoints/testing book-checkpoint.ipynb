# Testing new features

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import joblib
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns

# Reading datasets
def read_datasets():
    genuine_users = pd.read_csv("data/users.csv")
    fake_users = pd.read_csv("data/fusers.csv")
    x = pd.concat([genuine_users, fake_users], ignore_index=True)
    y = [0] * len(fake_users) + [1] * len(genuine_users)
    return x, y

# Predicting gender using gender_guesser
def predict_sex(name):
    import gender_guesser.detector as gender
    detector = gender.Detector()
    first_name = name.str.split(' ').str.get(0)
    sex = first_name.apply(lambda x: detector.get_gender(x).split()[0])
    sex_dict = {'female': -2, 'mostly_female': -1, 'unknown': 0, 'mostly_male': 1, 'male': 2}
    sex_code = sex.map(sex_dict).fillna(0).astype(int)
    return sex_code

# Extracting useful features for training
def extract_features(x):
    lang_dict = {name: i for i, name in enumerate(x['lang'].unique())}
    x['lang_code'] = x['lang'].map(lambda x: lang_dict.get(x, -1)).astype(int)
    x['sex_code'] = predict_sex(x['name'])
    feature_columns_to_use = [
        'statuses_count', 
        'followers_count', 
        'friends_count', 
        'favourites_count', 
        'listed_count', 
        'sex_code', 
        'lang_code',
        'name'
    ]
    missing_cols = [col for col in feature_columns_to_use if col not in x.columns]
    if missing_cols:
        raise ValueError(f"Required columns missing from DataFrame: {', '.join(missing_cols)}")
    x = x[feature_columns_to_use]
    return x

# Anomaly detection using Isolation Forest
def detect_anomalies(x, contamination=0.05):
    behavior_features = x[['statuses_count', 'followers_count', 'friends_count', 'favourites_count']]
    iso_forest = IsolationForest(contamination=contamination, random_state=42)
    anomaly_labels = iso_forest.fit_predict(behavior_features)
    x['anomaly'] = anomaly_labels
    num_anomalies = (anomaly_labels == -1).sum()
    
    # Filter for anomalous users
    anomalous_users = x[x['anomaly'] == -1]['name'].values
    print(f"Detected {num_anomalies} anomalies in user behavior with contamination level {contamination}.")
    print("Anomalous Users:", anomalous_users, "\n")

    return x, anomalous_users

# Visualize user behavior over time and highlight anomalies
def visualize_behavior_over_time(x, date_column='created_at', anomalous_users=None):
    if date_column not in x.columns:
        print(f"Column '{date_column}' not found in dataset.")
        return

    x[date_column] = pd.to_datetime(x[date_column], errors='coerce')
    x = x.dropna(subset=[date_column]).sort_values(by=date_column)

    # Separate anomalies from normal data
    normal_data = x[x['anomaly'] == 1]
    anomaly_data = x[x['anomaly'] == -1]

    plt.figure(figsize=(12, 6))

    # Plotting behavior over time
    plt.plot(normal_data[date_column], normal_data['followers_count'], label="Normal Followers Count", color='blue')
    plt.plot(normal_data[date_column], normal_data['statuses_count'], label="Normal Statuses Count", color='green')

    # Plotting anomalies
    plt.scatter(anomaly_data[date_column], anomaly_data['followers_count'], color='red', label='Anomaly (Followers)', s=50)
    plt.scatter(anomaly_data[date_column], anomaly_data['statuses_count'], color='orange', label='Anomaly (Statuses)', s=50)

    # Adding annotations for anomalous user names
    for i, row in anomaly_data.iterrows():
        plt.text(row[date_column], row['followers_count'], row['name'], fontsize=8, color='red')
    
    plt.xlabel('Date')
    plt.ylabel('Count')
    plt.title('User Behavior Over Time (with Anomalies Highlighted)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
def train_xgboost(X, y):
    # Drop non-numeric columns before scaling and model training
    X_numeric = X.drop(columns=['name'], errors='ignore')  # Exclude 'name' or any non-numeric columns
    
    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_numeric, y, test_size=0.20, random_state=42, stratify=y)

    # Scale the data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # XGBoost model setup
    xgb_model = xgb.XGBClassifier(objective='binary:logistic', eval_metric='logloss', random_state=42)
    
    # Grid search for hyperparameter tuning
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.8, 1.0]
    }
    
    grid_search = GridSearchCV(xgb_model, param_grid, cv=3, scoring='accuracy', verbose=1)
    grid_search.fit(X_train, y_train)

    print("Best parameters found: ", grid_search.best_params_)
    clf = grid_search.best_estimator_
    
    # Predictions and evaluation
    predictions = clf.predict(X_test)
    predicted_probabilities = clf.predict_proba(X_test)[:, 1]

    print("Classification Report (XGBoost):\n", classification_report(y_test, predictions))
    print("Confusion Matrix (XGBoost):\n", confusion_matrix(y_test, predictions))

    return y_test, predictions, predicted_probabilities, clf, X_train, X_test, scaler


# Save model and scaler
def save_model_and_scaler(model, scaler):
    model.save_model('xgboost_model.json')
    joblib.dump(scaler, 'scaler.pkl')

# Main function
def main():
    print("Reading datasets...\n")
    x, y = read_datasets()

    print("Extracting features...\n")
    x = extract_features(x)

    print("Detecting anomalies in user behavior...\n")
    contamination_level = 0.05  # Adjust this value as needed
    x, anomalous_users = detect_anomalies(x, contamination=contamination_level)

    print("Visualizing user behavior over time...\n")
    visualize_behavior_over_time(x, date_column='created_at', anomalous_users=anomalous_users)

    print("Training XGBoost model...\n")
    y_test_xgb, y_pred_xgb, y_pred_prob_xgb, model_xgb, X_train, X_test, scaler = train_xgboost(x, y)

    print("Saving model and scaler...\n")
    save_model_and_scaler(model_xgb, scaler)

    print("Evaluation metrics for XGBoost:\n")
    accuracy_xgb = (y_test_xgb == y_pred_xgb).mean()
    print(f'Classification Accuracy on Test dataset (XGBoost): {accuracy_xgb:.2f}')

    # Plot confusion matrix
    cm = confusion_matrix(y_test_xgb, y_pred_xgb)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Fake', 'Genuine'], yticklabels=['Fake', 'Genuine'])
    plt.title('Confusion Matrix (XGBoost)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

    # ROC curve
    false_positive_rate, true_positive_rate, _ = roc_curve(y_test_xgb, y_pred_prob_xgb)
    roc_auc = auc(false_positive_rate, true_positive_rate)

    plt.figure()
    plt.title('Receiver Operating Characteristic (ROC)')
    plt.plot(false_positive_rate, true_positive_rate, 'b', label='AUC = %0.2f' % roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.tight_layout()
    plt.show()

    # SHAP values for feature importance
    import shap
    explainer = shap.Explainer(model_xgb)
    shap_values = explainer(X_train)

    plt.figure()
    shap.summary_plot(shap_values, X_train)
    plt.show()

if __name__ == "__main__":
    main()
```
